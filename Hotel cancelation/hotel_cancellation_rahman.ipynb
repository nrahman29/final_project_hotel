```python
# get a list of the columns
col_list = list(df1)
# use this handy way to swap the elements
col_list[3], col_list[5] = col_list[5], col_list[3]
# assign back, the order will now be swapped
df1.columns = col_list
```

```python
df1.head()
```

```python
#Converting typo 'Veganuary' to 'January'.
df1['arrival_date_month'].replace('Veganuary', 'January', inplace=True)
```

```python
#Converting all negative values in the lead_time column to positive and rounding them.
df1['lead_time'] = df1['lead_time'].abs()
df1['lead_time'] = df1['lead_time'].round()
```

```python
#Converting all negative values in the stays_in_weekend_nights column and stays_in_week_nights to positive and rounding them.
df1['stays_in_weekend_nights'] = df1['stays_in_weekend_nights'].abs()
df1['stays_in_weekend_nights'] = df1['stays_in_weekend_nights'].round()

df1['stays_in_week_nights'] = df1['stays_in_week_nights'].abs()
df1['stays_in_week_nights'] = df1['stays_in_week_nights'].round()

```

```python
# removing meal and country columns
df1 = df1.drop(['meal','country'], axis = 1, index=None)
```
